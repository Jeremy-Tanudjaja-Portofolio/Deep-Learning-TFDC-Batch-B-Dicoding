Imports

```python
import numpy as np
import pandas as pd
import PIL as pil
import scipy as sc
import tensorflow as tf
import tensorflow_datasets as tfds
from sklearn.model_selection import train_test_split
```

Get the data

```python
bbc = pd.read_csv('https://github.com/dicodingacademy/assets/raw/main/Simulation/machine_learning/bbc-text.csv')
```

```python
bbc.head(5)
```

Split the data

```python
train_x, test_x, train_y, test_y = train_test_split(bbc['text'], bbc['category'], test_size = 0.2, train_size = 0.8, shuffle=False)
```

```python
train_x.head(5)
```

```python
test_x.head(5)
```

```python
train_y.head(5)
```

```python
test_y.head(5)
```

Preprocess the text

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer(oov_token='<OOV', num_words = 12000)
tokenizer.fit_on_texts(train_x)
```

```python
train_x_token = tokenizer.texts_to_sequences(train_x)
test_x_token = tokenizer.texts_to_sequences(test_x)
```

```python
train_x_token = tf.keras.preprocessing.sequence.pad_sequences(train_x_token, padding='post', maxlen=120, truncating='post')
test_x_token = tf.keras.preprocessing.sequence.pad_sequences(test_x_token, padding='post', maxlen=120, truncating='post')
```

```python
train_x_token = np.array(train_x_token)
test_x_token = np.array(test_x_token)
```

```python
train_x_token
```

```python
train_x_token.shape
```

```python
test_x_token.shape
```

Preprocess the Label

```python
label_tokenizer = tf.keras.preprocessing.text.Tokenizer()
label_tokenizer.fit_on_texts(train_y)
```

```python
train_y_token = np.array(label_tokenizer.texts_to_sequences(train_y))
test_y_token = np.array(label_tokenizer.texts_to_sequences(test_y))
```

```python
train_y_token[0:2]
```

Model Creation

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=12000, output_dim=16, input_length=120),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32, return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(6, activation='softmax')
])
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(train_x_token,train_y_token, validation_data=(test_x_token, test_y_token), epochs = 15)
```
