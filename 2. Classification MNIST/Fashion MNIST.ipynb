Imports

```python
import numpy as np
import pandas as pd
import PIL as pil
import scipy as sc
import tensorflow as tf
import tensorflow_datasets as tfds
```

Get the data

```python
data, info = tfds.load('fashion_mnist', split=['train','test'], with_info = True, as_supervised = True)
```

```python
train_data, test_data = data
```

```python
train_x = []
train_y = []
test_x = []
test_y = []
```

```python
for image, label in train_data:
  image = tf.image.resize(image,(28,28))
  train_x.append(image.numpy())
  train_y.append(label.numpy())
```

```python
for image, label in test_data:
  image = tf.image.resize(image,(28,28))
  test_x.append(image.numpy())
  test_y.append(label.numpy())
```

```python
train_x = np.array(train_x)
train_y = np.array(train_y)

test_x = np.array(test_x)
test_y = np.array(test_y)
```

```python
train_image_data = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255.)
train_gen = train_image_data.flow(train_x, train_y)
```

```python
test_image_data = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255.)
test_gen = test_image_data.flow(test_x, test_y)
```

Model

```python
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(28,28,1)),
    tf.keras.layers.Conv2D(32,(3,3),activation='relu'),
    tf.keras.layers.MaxPooling2D(2),
    tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
```

```python
model.fit(train_gen, validation_data=test_gen, epochs =15)
```
